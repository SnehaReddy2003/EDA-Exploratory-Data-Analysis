## 1.What are Embeddings?

- Embeddings are numercical representation of text, images, audio, vedios of any data.For text (usually our focus ), an embedding is:
    -A vector (a long list of numbers)
    - Generated by a model
    - Designed so that simialr meanings -similar numbers
    
## 2.Why Do we Use Embeddings?
- Becoause computers dont understand text dierctly - they understand numbers
- Embeddings allow systems to:
    - Search semantically(meaning-based search)
    - Build recommendations system
    - Measure similarity between tetxs
    - Clusters documents
    - RAG (Retrieval- Agrumentes Geneartion),including ChatGroq Workflows
    
### 3. What is a Similarity Score?
- Once you have embeddings you measure similarity using a mathematical formula
- The most common one is:
- **Cosine Similarity**:
    -It measures the angle between two vectors
        - 1.0-->Identical meaning
        - 0.0-->unrelated
        - (-)1.0-->Opposite meaning
